```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('IMDb Movie Database.csv')

# Assuming 'df' is your DataFrame and 'IMDb Score (1-10)' is the column containing IMDb scores
top_10_movies = df.nlargest(10, 'IMDb Score (1-10)')

# Plotting the column chart
plt.figure(figsize=(32, 12))
barplot = sns.barplot(x='Title', y='IMDb Score (1-10)', data=top_10_movies, palette='dark')
plt.title('Top 10 Movies by IMDb Score')
plt.xlabel('Title')
plt.ylabel('IMDb Score (1-10)')

# Adding IMDb scores as text labels on the bars
for index, value in enumerate(top_10_movies['IMDb Score (1-10)']):
    barplot.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=15, color='black')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Country': ['Philippines', 'Afghanistan', 'Indonesia', 'Poland', 'Colombia',
                     'Israel', 'Finland', 'Argentina', 'Romania', 'Iceland'],
        'Total Budget': [7000, 46000, 1100000, 2400000, 3000000, 3450000, 3850000,
                         4900000, 12590000, 13800000]}

df = pd.DataFrame(data)

# Convert 'Total Budget' to millions for better readability
df['Total Budget (Millions)'] = df['Total Budget'] / 1e6

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('yellow')  # Set background color

plt.plot(df['Country'], df['Total Budget (Millions)'], marker='o', linestyle='-', color='b')

# Adding labels and title
plt.title('CountryWise Low budget movie')
plt.xlabel('Country')
plt.ylabel('Total Budget (Millions)')

# Adding data labels
for i, txt in enumerate(df['Total Budget (Millions)']):
    plt.annotate(f'{txt:.2f}M', (df['Country'][i], df['Total Budget (Millions)'][i]),
                 textcoords="offset points", xytext=(0, 5), ha='center')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd

# Load the dataset
movie_data = pd.read_csv('IMDb Movie Database.csv')

# Display the first few rows to understand the data
print(movie_data.head())
```

```python
# Calculate profitability
movie_data['Profitability'] = movie_data['Gross Revenue'] - movie_data['Budget']

```

```python
actor_profitability = movie_data.groupby('Lead Actor')['Profitability'].sum().reset_index()
```

```python
ranked_actors = actor_profitability.sort_values(by='Profitability', ascending=False)
```

```python
ranked_actors
```

```python
import matplotlib.pyplot as plt

# Assuming you have already calculated ranked_actors

# Plotting the top 10 most profitable actors as a donut chart with a legend
top_actors = ranked_actors.head(10)

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_actors['Profitability'], labels=top_actors['Lead Actor'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))

# Draw a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.title('Top 10 Most Profitable Actors')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('IMDb Movie Database.csv')
data = pd.DataFrame(data)
genre_revenue = data.groupby('Genre')['Gross Revenue'].sum().reset_index()
top_genres = genre_revenue.sort_values(by='Gross Revenue', ascending=False).head(5)
top_genres
```

```python

from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(10, 6))
bars = plt.bar(top_genres['Genre'], top_genres['Gross Revenue'], color='green')
plt.xlabel('Genre')
plt.ylabel('Gross Revenue (Billions)')
plt.title('Top 5 Genres by Gross Revenue')

# Format y-axis labels in billions using FuncFormatter
formatter = FuncFormatter(lambda x, _: f'${int(x/1e9)}B')
plt.gca().yaxis.set_major_formatter(formatter)

# Annotate each bar with its value above the bar
for bar, value in zip(bars, top_genres['Gross Revenue']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05e9, f'${int(value/1e9)}B', ha='center', va='bottom', color='black')

plt.show()
```

```python
import pandas as pd

# Load your IMDb dataset
data = pd.read_csv('IMDb Movie Database.csv')

# Ensure that 'budget' and 'gross revenue' columns are numeric
data['Budget'] = pd.to_numeric(data['Budget'], errors='coerce')
data['Gross Revenue'] = pd.to_numeric(data['Gross Revenue'], errors='coerce')

# Calculate profitability
data['profitability'] = (data['Gross Revenue'] - data['Budget']) / data['Budget']

# Sort by profitability in ascending order to get the flop movies
flop_movies = data.sort_values(by='profitability').head(10)

# Display the top 10 flop movies
print(flop_movies[['Title', 'Budget', 'Gross Revenue', 'profitability']])
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot 'Budget' on the primary y-axis as a column chart
ax1.bar(flop_movies['Title'], flop_movies['Budget'] / 1000, color='b', alpha=0.7, label='Budget (in thousands)')
ax1.set_xlabel('Movie Title')
ax1.set_ylabel('Budget (in thousands)', color='b')
ax1.tick_params('y', colors='b')
ax1.set_xticklabels(flop_movies['Title'], rotation=45, ha='right')

# Create a secondary y-axis for 'Gross Revenue' as a line chart
ax2 = ax1.twinx()
ax2.plot(flop_movies['Title'], flop_movies['Gross Revenue'] / 1000, color='r', marker='o', label='Gross Revenue (in thousands)')
ax2.set_ylabel('Gross Revenue (in thousands)', color='r')
ax2.tick_params('y', colors='r')

# Title and legend
plt.title('Top 10 Flop Movies: Budget (Column) and Gross Revenue (Line)')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```
