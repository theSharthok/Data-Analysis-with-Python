convert overs in numeric value

```python
import pandas as pd

data = {'overs': df['overs'],
        'wd':df['wd'],
        'nb':df['nb']}
df1 = pd.DataFrame(data)

# Define the formula function
def custom_formula(overs, wd, nb):
    return (int(overs) * 6) + (overs % 1 * 10) +(wd +nb)

# Apply the formula to the 'FloatColumn' and create a new 'NewColumn'
df1['bowls'] = df.apply(lambda row: custom_formula(row['overs'], row['wd'], row['nb']), axis=1).astype('int64')

# Display the DataFrame with the new column

print("\nDataFrame with the new column:")
print(df1[ 'bowls'])
```

```python
df1
```

```python
df = pd.concat([df,df1['bowls']],axis=1)
```

playerwise bowls

```python
playerwise_bowls = df.groupby(['player', 'team'])['bowls'].sum().reset_index()
print("Playerwise Bowls:")
print(playerwise_bowls)
```

top players to bowls max and min

```python
# Sort the dataframe by 'bowls' in descending order
sorted_playerwise_bowls = playerwise_bowls.sort_values(by='bowls', ascending=False)

# Top 10 bowlers who bowled the maximum
top_10_max = sorted_playerwise_bowls.head(10)

# Top 10 bowlers who bowled the minimum
top_10_min = sorted_playerwise_bowls.tail(10)

top_10_max , top_10_min
```

most dot balls by player(% of total bowls)

```python
import pandas as pd
import numpy as np

playerwise_stats = df.groupby(['player', 'team']).agg({'bowls': 'sum', 'dots': 'sum'}).reset_index()
playerwise_stats['dot_percentage'] = (playerwise_stats['dots'] / playerwise_stats['bowls']) * 100
sorted_playerwise_stats = playerwise_stats.sort_values(by='dots', ascending=False)
top_10_max = sorted_playerwise_stats.head(10).copy()  # Explicitly create a copy

import warnings

# Suppress division by zero warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)

# Calculate dot percentage of total bowls for top 10
top_10_max.loc[:, 'dot_percentage_of_total_bowls'] = np.where(top_10_max['bowls'] != 0,
                                                               (top_10_max['dots'] / top_10_max['bowls']) * 100,
                                                               0)
# Round values to two decimal places
top_10_max = top_10_max.round(2)

# Display the results
print("Top 10 Players (Dots Percentage):")
print(top_10_max[['player', 'team', 'dots', 'bowls', 'dot_percentage']])
```

extra bowls player based on (wide + no) bowls

```python
import pandas as pd

df['extra_bowls'] = df['wd'] + df['nb']

# Group by 'player' and calculate the sum of 'extra_bowls'
player_extra_bowls = df.groupby('player')['extra_bowls'].sum().reset_index()

# Sort the DataFrame by the sum in descending order
top_20_players = player_extra_bowls.sort_values(by='extra_bowls', ascending=False).head(20)

# Merge with the original DataFrame to get total bowls for each selected player
top_20_players_with_total_bowls = pd.merge(top_20_players, df.groupby('player')['bowls'].sum().reset_index(), on='player')

print("Top players most extra bowls:\n")
print(top_20_players_with_total_bowls)
```

teamwise total runs concede

```python
teamwise_total_runs_conceded = df.groupby('team')['runs'].sum().reset_index()
teamwise_total_runs_conceded = teamwise_total_runs_conceded.sort_values(by='runs', ascending=False)

print("Teamwise Total Runs Conceded:\n")
print(teamwise_total_runs_conceded)
```

boundary concede by team

```python
# Group by 'team' and calculate the total boundaries (4s + 6s) for each team
teamwise_boundaries = df.groupby('team').agg({'4s': 'sum', '6s': 'sum'}).reset_index()
teamwise_boundaries['boundaries_conceded'] = teamwise_boundaries['4s'] + teamwise_boundaries['6s']

# Sort the DataFrame in descending order based on 'total_boundaries'
teamwise_boundaries = teamwise_boundaries.sort_values(by='boundaries_conceded', ascending=False)

# Display the result
print("Teamwise Total Boundaries Conceded:\n")
print(teamwise_boundaries)
```

inningswise boundaries

```python
# Group by 'innings' and calculate the total boundaries (4s + 6s) for each innings
inningswise_boundaries = df.groupby('innings').agg({'4s': 'sum', '6s': 'sum'}).reset_index()
inningswise_boundaries['total_boundaries'] = inningswise_boundaries['4s'] + inningswise_boundaries['6s']

import matplotlib.pyplot as plt

# Plot a pie chart with labeled slices
labels = ['Innings 1', 'Innings 2']
sizes = inningswise_boundaries['total_boundaries']
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Innings-wise Total Boundaries Conceded')
plt.show()
```

wickets taken by team

```python
# Group by 'team' and calculate the total wickets taken for each team
teamwise_wickets = df.groupby('team')['wickets'].sum().reset_index()

# Sort the DataFrame in descending order based on 'wickets'
teamwise_wickets = teamwise_wickets.sort_values(by='wickets', ascending=False)

# Display the result
print("Wickets Taken by Team:")
print(teamwise_wickets)
```

inningswise wickets taken

```python
# Group by 'innings' and calculate the total wickets taken for each inning
innings_wickets = df.groupby('innings')['wickets'].sum().reset_index()

# Plot a pie chart
labels = ['Innings 1', 'Innings 2']
sizes = innings_wickets['wickets']
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Wickets Taken in Innings 1 and Innings 2')
plt.show()
```
