goals by each team

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
data = {'team1': df['team1'],
        'team2': df['team2'],
        'number of goals team1': df['number of goals team1'],
        'number of goals team2': df['number of goals team2']}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total goals for each team
team1_total_goals = df.groupby('team1')['number of goals team1'].sum().reset_index()
team2_total_goals = df.groupby('team2')['number of goals team2'].sum().reset_index()

# Merge the results into a single DataFrame
total_goals_by_team = pd.merge(team1_total_goals, team2_total_goals, left_on='team1', right_on='team2', how='outer')

# Fill NaN values with 0
total_goals_by_team.fillna(0, inplace=True)

# Sum the total goals for each team
total_goals_by_team['total_goals'] = total_goals_by_team['number of goals team1'] + total_goals_by_team['number of goals team2']


# Rename columns for clarity
total_goals_by_team.rename(columns={'team1': 'team', 'total_goals': 'total goals'}, inplace=True)

# Sort the DataFrame by total goals in descending order
total_goals_by_team = total_goals_by_team.sort_values(by='total goals', ascending=False)
total_goals_by_team
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
data = {'team1': df['team1'],
        'team2': df['team2'],
        'number of goals team1': df['number of goals team1'],
        'number of goals team2': df['number of goals team2']}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total goals for each team
team1_total_goals = df.groupby('team1')['number of goals team1'].sum().reset_index()
team2_total_goals = df.groupby('team2')['number of goals team2'].sum().reset_index()

# Merge the results into a single DataFrame
total_goals_by_team = pd.merge(team1_total_goals, team2_total_goals, left_on='team1', right_on='team2', how='outer')

# Fill NaN values with 0
total_goals_by_team.fillna(0, inplace=True)

# Sum the total goals for each team
total_goals_by_team['total_goals'] = total_goals_by_team['number of goals team1'] + total_goals_by_team['number of goals team2']


# Rename columns for clarity
total_goals_by_team.rename(columns={'team1': 'team', 'total_goals': 'total goals'}, inplace=True)

# Sort the DataFrame by total goals in descending order
total_goals_by_team = total_goals_by_team.sort_values(by='total goals', ascending=False)

# Select the top 10 teams based on total goals
top_10_teams = total_goals_by_team.head(10)

# Define different colors for each bar
colors = ['skyblue', 'lightcoral', 'lightgreen', 'orange', 'lightblue', 'pink', 'yellow', 'lightgray', 'purple', 'brown']

# Plot the histogram for the top 10 teams with different colors
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_teams['team'], top_10_teams['total goals'], color=colors)
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.title('Top Teams by Total Goals')
plt.xticks(rotation=45, ha='right')

# Add data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.show()
```

yellow cards and red cards conceded by team

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
data = {'team1': df['team1'],
        'team2': df['team2'],
        'yellow cards team1': df['yellow cards team1'],
        'yellow cards team2': df['yellow cards team2'],
        'red cards team1': df['red cards team1'],
        'red cards team2': df['red cards team2']}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total yellow cards for each team
team1_total_yellow_cards = df.groupby('team1')['yellow cards team1'].sum().reset_index()
team2_total_yellow_cards = df.groupby('team2')['yellow cards team2'].sum().reset_index()

# Calculate total red cards for each team
team1_total_red_cards = df.groupby('team1')['red cards team1'].sum().reset_index()
team2_total_red_cards = df.groupby('team2')['red cards team2'].sum().reset_index()

# Merge the results into a single DataFrame for yellow cards
total_yellow_cards_by_team = pd.merge(team1_total_yellow_cards, team2_total_yellow_cards, left_on='team1', right_on='team2', how='outer')
total_yellow_cards_by_team.fillna(0, inplace=True)
total_yellow_cards_by_team['total_yellow_cards'] = total_yellow_cards_by_team['yellow cards team1'] + total_yellow_cards_by_team['yellow cards team2']

# Merge the results into a single DataFrame for red cards
total_red_cards_by_team = pd.merge(team1_total_red_cards, team2_total_red_cards, left_on='team1', right_on='team2', how='outer')
total_red_cards_by_team.fillna(0, inplace=True)
total_red_cards_by_team['total_red_cards'] = total_red_cards_by_team['red cards team1'] + total_red_cards_by_team['red cards team2']

# Rename columns for clarity
total_yellow_cards_by_team.rename(columns={'team1': 'team', 'total_yellow_cards': 'total yellow cards'}, inplace=True)
total_red_cards_by_team.rename(columns={'team1': 'team', 'total_red_cards': 'total red cards'}, inplace=True)

# Display the results
print("Team-wise Total Yellow Cards:")
print(total_yellow_cards_by_team)

print("\nTeam-wise Total Red Cards:")
print(total_red_cards_by_team)
```

goal preventions and own goals

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
data = {'team1': df['team1'],
        'team2': df['team2'],
        'goal preventions team1': df['goal preventions team1'],
        'goal preventions team2': df['goal preventions team2'],
        'own goals team1': df['own goals team1'],
        'own goals team2': df['own goals team2']}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate total goal preventions for each team
team1_total_preventions = df.groupby('team1')['goal preventions team1'].sum().reset_index()
team2_total_preventions = df.groupby('team2')['goal preventions team2'].sum().reset_index()

# Calculate total own goals for each team
team1_total_own_goals = df.groupby('team1')['own goals team1'].sum().reset_index()
team2_total_own_goals = df.groupby('team2')['own goals team2'].sum().reset_index()

# Merge the results into a single DataFrame for goal preventions
total_preventions_by_team = pd.merge(team1_total_preventions, team2_total_preventions, left_on='team1', right_on='team2', how='outer')
total_preventions_by_team.fillna(0, inplace=True)
total_preventions_by_team['total_preventions'] = total_preventions_by_team['goal preventions team1'] + total_preventions_by_team['goal preventions team2']

total_preventions_by_team.rename(columns={'team1': 'team', 'total_preventions': 'total preventions'}, inplace=True)

# Merge the results into a single DataFrame for own goals
total_own_goals_by_team = pd.merge(team1_total_own_goals, team2_total_own_goals, left_on='team1', right_on='team2', how='outer')
total_own_goals_by_team.fillna(0, inplace=True)
total_own_goals_by_team['total_own_goals'] = total_own_goals_by_team['own goals team1'] + total_own_goals_by_team['own goals team2']

total_own_goals_by_team.rename(columns={'team1': 'team', 'total_own_goals': 'total own goals'}, inplace=True)

# Display the results
print("Total Preventions by Team:")
print(total_preventions_by_team)
print("\nTotal Own Goals by Team:")
print(total_own_goals_by_team)
```
